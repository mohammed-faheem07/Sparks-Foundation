## Importing Required Libraries

## Checking the relation of each column with other column

```python
co_relation = csv_file.corr()
sns.heatmap(co_relation, xticklabels=co_relation.columns, yticklabels=co_relation.columns,
           annot=True)
```

## Analyzing the sales of different categorys

```python
plt.bar(csv_file.Category,csv_file.Sales)
plt.xlabel('Category')
plt.ylabel('Sales')
plt.grid()
plt.show()
```

## Analyzing the sales of different ranges

```python
plt.bar(csv_file.Region,csv_file.Sales)
plt.xlabel('Region')
plt.ylabel('sales')
plt.grid()
plt.show()
```

## Analyzing the profit of different segments

```python
plt.bar(csv_file.Segment,csv_file.Profit)
plt.xlabel('Segment')
plt.ylabel('Profit')
plt.xticks(rotation=90,fontsize=6)
plt.grid()
plt.show()
```

## Analyzing the profit of different regions

```python
plt.bar(csv_file.Region,csv_file.Profit)
plt.xlabel('Region')
plt.ylabel('Profit')
plt.grid()
plt.show()
```

## Analyzing the profit of different Categorys

```python
plt.bar(csv_file.Category,csv_file.Profit)
plt.xlabel('Category')
plt.ylabel('Profit')
plt.grid()
plt.show()
```

## Analyzing the profit of different sub-categorys

```python
plt.bar(csv_file['Sub-Category'],csv_file.Profit)
plt.xlabel('Sub-Category')
plt.ylabel('Profit')
plt.xticks(rotation=90,fontsize=6)
plt.grid()
plt.show()
```

## Analyzing the profit of different ship modes

```python
plt.bar(csv_file['Ship Mode'],csv_file.Profit)
plt.xlabel('Ship_Mode')
plt.ylabel('Profit')
plt.grid()
plt.show()
```
